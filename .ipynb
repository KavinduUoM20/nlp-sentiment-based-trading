```python
def check_duplicates(df, drop=False):
    """
    Check for duplicate rows in a DataFrame, print examples, and optionally drop them.

    Parameters:
    df (pd.DataFrame): The DataFrame to check for duplicates.
    drop (bool): Whether to drop duplicate rows. Default is False.

    Returns:
    pd.DataFrame: DataFrame with duplicates checked (and possibly dropped).
    """
    num_duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {num_duplicates}")

    if num_duplicates > 0:
        # Print examples of duplicate rows
        duplicate_rows = df[df.duplicated(keep=False)]
        print("Examples of duplicate rows:")
        display(duplicate_rows.head(2))

    if drop:
        df = df.drop_duplicates()
        print("Duplicate rows dropped.")

    return df

for name, df in dataframes.items():
    print(f"Checking duplicates in {name}:")
    dataframes[name] = check_duplicates(df, drop=True)
    print("\n")

# Check the dataframes after dropping duplicates
for name, df in dataframes.items():
    print(f"{name} after dropping duplicates:")
    display(df.head())
    print("\n")
```

# **Preprocessing**

```python
def cleaning( review, remove_stopwords=True):
    review_text = re.sub("[^a-zA-Z]"," ", review)
    words = review_text.lower().split()
    if remove_stopwords:
        stops = set(stopwords.words("english"))
        words = [w for w in words if not w in stops]
    b=[]
    stemmer = english_stemmer
    for word in words:
        b.append(stemmer.stem(word))
    return(words)
```

```python
clean_Text = []
for review in news_df['Abstract']:
    clean_Text.append( " ".join(cleaning(str(review))))
```

```python
Top_Words_Review =pd.Series(' '.join(clean_Text).lower().split()).value_counts()[:20]
print ("Top Count Words Used In Review", Top_Words_Review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df=4, max_features = 10000)
vz = vectorizer.fit_transform(clean_Text)
tfidf = dict(zip(vectorizer.get_feature_names_out(), vectorizer.idf_))
print("Buy: " + str(tfidf["buy"]))
print("Like: " + str(tfidf["like"]))
print("Sell: " + str(tfidf["sell"]))
print("Drops: " + str(tfidf["drop"]))
print("Share: " + str(tfidf["share"]))
print("growth: " + str(tfidf["growth"]))
print("profit: " + str(tfidf["profit"]))
```

# Sentiment Analysis using nltk SentimentIntensityAnalyser

The polarity score returns 4 values
1. neutral
2. negative
3. Positive
4. compound

compound is overall **sentiment** of the **text** entered inside the **polarity score....**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
import csv

file = open('datapolarity.csv','w',encoding='ISO-8859-1')
file.write("Cleaned Title,Compound,Negative,Neutral,Positive\n")
Senti = SentimentIntensityAnalyzer()
sample_review = clean_Text
for sentence in sample_review:
    ss = Senti.polarity_scores(sentence)
    file.write(sentence+","+str(ss['compound'])+","+str(ss['neg'])+","+str(ss['neu'])+","+str(ss['pos'])+"\n")
file.close()
```

```python
# Preprocess and aggregate function for tweets and news
def aggregate(df, text_column):
    df['date'] = pd.to_datetime(df['date'])
    # Group by date and concatenate texts
    df_aggregated = df.groupby('date')[text_column].apply(lambda x: ' '.join(x)).reset_index()
    return df_aggregated

aggregated_tweets = aggregate(tweets_df, 'tweet')
aggregated_news = aggregate(news_df, 'Abstract')
```
